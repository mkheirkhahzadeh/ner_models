### Create en_dataset with English rows from three datasets (train, validation, test)
We find English indexes in each dataset and return the filtered dataset  

```python
from datasets import DatasetDict
import time
#Call filter for each element which is a datasetDict

start_time = time.time()

en_train_ds = data['train'].filter(lambda r: r['lang']== 'en')
print("---  train: ---",  (time.time() - start_time))

start_time = time.time()

en_val_ds = data['validation'].filter(lambda r: r['lang']== 'en')
print("---  validation ---" , (time.time() - start_time))

start_time = time.time()

en_test_ds = data['test'].filter(lambda r: r['lang']== 'en')
print("--- test: ---" , (time.time() - start_time))



#STEP#
en_dataset = DatasetDict({'train':en_train_ds, 'validation':en_val_ds,'test':en_test_ds })
```

```python
en_dataset
```

```python
en_dataset.push_to_hub('MaryDatascientist/en_dataset')
```
